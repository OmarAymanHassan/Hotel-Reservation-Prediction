## Using colTransformer 

```python
from sklearn.preprocessing import StandardScaler ,OneHotEncoder,OrdinalEncoder , FunctionTransformer
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

numeric_transfomer_stasca =StandardScaler()
cat_transformation= OrdinalEncoder(handle_unknown="use_encoded_value" , unknown_value=-1)
#cat_transformation = OneHotEncoder()
skew_transormation = FunctionTransformer(np.log1p, validate=True)

#pre_processor = ColumnTransformer(transformers=[("num_encoding" , numeric_transformer_ordinal , num_cols) , ("standard_scaling" , numeric_transfomer_stasca , num_cols) , ("cat_encoding" , cat_transformation , cat_cols)])
#pre_processor = Pipeline(steps=[("num_encoding" , numeric_transformer_ordinal , num_cols) , ("standard_scaling" , numeric_transfomer_stasca , num_cols) , ("cat_encoding" , cat_transformation , cat_cols)])
skew_pipeline = Pipeline(steps=[("skew_transformation" , skew_transormation) , ("Scaling" , numeric_transfomer_stasca)])
numerical_pipeline = Pipeline(steps=[("standard_scaling" , numeric_transfomer_stasca)])
cat_pipeline = Pipeline(steps=[("ordinal_encoding" , cat_transformation)])

col_transfomer = ColumnTransformer([("skewness_transformation" , skew_pipeline , skewness_cols),("non_skewness_cols" , numerical_pipeline , non_skewness_cols) , ("cat_transformation" , cat_pipeline , cat_cols)], remainder="drop")
```

```python
from imblearn.pipeline import Pipeline as ImbPipeline
from sklearn.ensemble import RandomForestClassifier
from imblearn.over_sampling import SMOTE

full_transformation = ImbPipeline(steps=[("processing_all_cols" , col_transfomer) , ("SMOTE" , SMOTE(random_state=42))])
```
